```python
# WAP to accept two numbers from the user and display their sum, difference, product, and division.

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Sum = {num1 + num2}")
print(f"Difference = {num1 - num2}")
print(f"Product = {num1 * num2}")
print(f"Division = {num1 / num2}")
```

```python
# WAP to check whether a number entered by the user is even or odd.

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Number is Even")
else:
    print("Number is Odd")
```

```python
# WAP to find the largest among three numbers entered by the user.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print(f"Largest number is {a}")
elif b >= a and b >= c:
    print(f"Largest number is {b}")
else:
    print(f"Largest number is {c}")
```

```python
# WAP to print the multiplication table of a number entered by the user.

num = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# WAP to print numbers from 1 to 10 using a while loop.

i = 1

while i <= 10:
    print(i)
    i += 1
```

```python
# WAP to ask the user to enter names of thier 3 favorite movies and store them in a list.
fav_mov= []
print("Enter Your 3 favorite movie names: ")
for i in range(3):
    fav_mov.append(input(f"Your faviourite movie {i+1}: "))

print(f"The list of favorite movies is as follows: {fav_mov}")
```

```python
# WAP to display all elements of a list using a loop.

numbers = [10, 20, 30, 40, 50]

for num in numbers:
    print(num)
```

```python
# WAP to count the number of characters in a string entered by the user.

text = input("Enter a string: ")

print(f"The entered string is {text}.")
print(f"Number of characters = {len(text)}")
```

```python
# WAP to print the following pattern:
# *
# **
# ***
# ****
# *****

for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
```
