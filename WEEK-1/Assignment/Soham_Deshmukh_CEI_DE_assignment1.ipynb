```python
#cleaning the final_price column for further EDA and analysis
df["final_price"] = (
df["final_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
)

df["final_price"] = pd.to_numeric(
    df["final_price"],
    errors="coerce"
)
```

```python
print("Updated Data Type:")
print(df["final_price"].dtype)
```

```python
print("Values After Conversion:")
print(df["final_price"].head())
```

```python
#filling missing values in discount and final_column with thier resp. median values
df["discount"] = df["discount"].fillna(
    df["discount"].median()
)

df["final_price"] = df["final_price"].fillna(
    df["final_price"].median()
)
```

```python
#checking for missing values in the dataset
print(df.isnull().sum())
```

```python
#similarly handling missing values in remaing columns
df["seller_name"] = df["seller_name"].fillna(
    "Unknown Seller"
)

df["seller_information"] = df["seller_information"].fillna(
    "Information Not Available"
)

df["what_customers_said"] = df["what_customers_said"].fillna(
    "No Review Available"
)

df["videos"] = df["videos"].fillna(
    "No Video"
)

df["variations"] = df["variations"].fillna(
    "No Variations"
)
```

```python
#checking for missing values in the dataset
print(df.isnull().sum())
```

### Missing Value Handling

The dataset contained missing values in the following columns:

- what_customers_said
- seller_name
- seller_information
- videos
- variations

Since these columns contain textual information, missing values were replaced with meaningful placeholder values such as:

- "Unknown Seller"
- "No Review Available"
- "No Video"
- "No Variations"

This approach preserves all records while ensuring that no missing values remain in the dataset.

### Derived Column Creation

The assignment example suggests creating:

total_amount = price × quantity

However, the provided dataset does not contain a quantity-related field.

Therefore, a business-relevant derived column named
`discount_amount` was created using available attributes:

discount_amount = initial_price − final_price

This derived feature represents the monetary discount offered on
each product.

```python
#derieved column 1: Discount Amount
df["discount_amount"] = (
    df["initial_price"]
    - df["final_price"]
)
```

```python
df[
    [
        "initial_price",
        "final_price",
        "discount_amount"
    ]
].head()
```

```python
#derieved column 2: Discount Percentage
df["discount_percentage"] = (
    (df["initial_price"] - df["final_price"])
    / df["initial_price"]
) * 100
```

```python
df[
    [
        "initial_price",
        "final_price",
        "discount_percentage"
    ]
].head()
```

### Derived Column: Discount Percentage

A new column named `discount_percentage` was created to determine
the percentage discount offered on each product.

Formula:

discount_percentage =
((initial_price - final_price) / initial_price) × 100

This feature helps identify products offering the highest relative
discounts.

```python
#derived column 3: Popularity Score
df["popularity_score"] = (
    df["rating"]
    * np.log1p(df["ratings_count"])
)
```

```python
df["popularity_score"] = (
    df["rating"]
    * np.log1p(df["ratings_count"])
)
```

### Derived Column: Popularity Score

A popularity score was calculated using product ratings and the
number of customer ratings.

Formula:

popularity_score =
rating × log(1 + ratings_count)

This metric provides a better estimate of product popularity than
using ratings alone because it considers both rating quality and
customer engagement.
