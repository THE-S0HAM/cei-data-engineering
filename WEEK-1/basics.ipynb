```python
#Arithmetic Operators
a= 5
b= 10
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
```

```python
# Relational Operators
a = 5
b = 10
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
```

```python
# Assignment Operators
x = 5
print("x =", x)
x += 2
print("x += 2:", x)
x -= 1
print("x -= 1:", x)
x *= 3
print("x *= 3:", x)
x /= 2
print("x /= 2:", x)
x %= 4
print("x %= 4:", x)
x **= 2
print("x **= 2:", x)
```

```python
# Logical Operators
p = True
q = False
print("p and q:", p and q)
print("p or q:", p or q)
print("not p:", not p)
print("not q:", not q)
print("(a < b) and (x > 0):", (a < b) and (x > 0))
print("(a > b) or (x == 4):", (a > b) or (x == 4))
```

```python
# Control Flow
age = 18
if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")

score = 75
if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"
print("Grade:", grade)
```

```python
# Loops
for i in range(1, 6):
    print("for loop:", i)

count = 0
while count < 5:
    print("while loop:", count)
    count += 1

for num in range(1, 10):
    if num == 5:
        print("break at", num)
        break
    if num % 2 == 0:
        print("continue at", num)
        continue
    print("odd number:", num)
```
